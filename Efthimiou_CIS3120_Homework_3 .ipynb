2.4 Based on the search results from #2.3, find each user's user id, user name, posted tweet text and user profile image url. (10 points)

```python
#2.4
users = {u['id']:u for u in response2.includes['users']}

for tweet in response2.data:
    if users[tweet.author_id]: 
        user = users[tweet.author_id]
        print(user.id) #user id
        print(user.name) #user's name
        print(tweet.text) #tweet text
        print(user.username) #username
        print(user.profile_image_url) #user profile image
```
